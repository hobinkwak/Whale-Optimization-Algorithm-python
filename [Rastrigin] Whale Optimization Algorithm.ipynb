## Import whale optimization algorithm

```python
from woa import WOA
import numpy as np
import matplotlib.pyplot as plt
```

## Visualize Search Space

## Define Obj Function

```python
# minimum solution : (0, 0)
def f(X):
    A = 10
    sol = []
    for ind in X:
        sol.append(A*len(ind) + sum([(i**2 - A * np.cos(2 * np.pi * i)) for i in ind]) )
    return np.array(sol)
```

## Define Variable Boundary

```python
x_lb = y_lb = -5.12
x_ub = y_ub = 5.12

lb = np.array([x_lb, y_lb])
ub = np.array([x_ub, y_ub])
```

## Optimize

```python
woa = WOA(f, 100, 0.5, 100, lb, ub)
%time f_values, optimal_x = woa.run()

plt.plot(f_values)
plt.title(f"Optimal X : {optimal_x}")
```
