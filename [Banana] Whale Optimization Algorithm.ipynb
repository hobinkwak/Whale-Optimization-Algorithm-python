## Import whale optimization algorithm

```python
from woa import WOA
import numpy as np
import matplotlib.pyplot as plt
```

## Visualize Search Space

## Define Obj Function

```python
# minimum solution : (1, 1)
def f(X):
    a, b = 1, 100
    sol = []
    for x in X:
        sol.append(((a - x[0]) ** 2) + b*(x[1]-x[0]**2)**2)
    return np.array(sol)
```

## Define Variable Boundary

```python
x_lb = y_lb = -5
x_ub = y_ub = 5

lb = np.array([x_lb, y_lb])
ub = np.array([x_ub, y_ub])
```

## Optimize

```python
woa = WOA(f, 100, 0.5, 100, lb, ub)
%time f_values, optimal_x = woa.run()

plt.plot(f_values)
plt.title(f"Optimal X : {optimal_x}")
```
